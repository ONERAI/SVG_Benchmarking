```r
locs=cbind(rep(0:50, 51), rep(0:50, each=51))
```

```r
V=cov.matern(as.matrix(dist(locs)), nu=1/2, alpha = 10)
```

```r
set.seed(20)
z=mvrnorm(mu=rep(0, 51^2), Sigma=V)
z=matrix(z, ncol=51)
persp(x=(0:50), y=(0:50), z,  theta=45, phi=35, r=5, expand=0.6, axes=T,
ticktype="detailed", xlab="x", ylab="y", zlab="z")
filled.contour(x=0:50, y=0:50, z, color.palette=gray.colors)
```
