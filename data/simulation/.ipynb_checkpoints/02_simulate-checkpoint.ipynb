```r
sim.groups <- splatSimulate(group.prob = c(0.5, 0.2, 0.2, 0.1), method = "groups",
                            verbose = FALSE, nGenes=10000)
sim.groups <- logNormCounts(sim.groups)
sim.groups <- runPCA(sim.groups)
plotPCA(sim.groups, colour_by = "Group")
```
