```python
import numpy as np
import pandas as pd
import scipy as sp
import scanpy as sc
import squidpy as sq
import anndata as ad
import simstpy as sim
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for sigma in [0.5, 1.0, 1.5, 2.0, 2.5, 3.0]:
    for alpha in [0.0]:
        adata = sc.read_h5ad(f"./Covariance_v2/sigma_{sigma}_alpha_{alpha}.h5ad")
        sq.pl.spatial_scatter(adata, shape=None, color=adata.var_names[adata.var.spatially_variable][65], frameon=False,
                              dpi=60, size=120, title=f"sigma_{sigma}_alpha_{alpha}")
```
